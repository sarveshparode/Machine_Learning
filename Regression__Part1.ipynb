Regression-- It is method for understanding relationship between dependent and independent variables --It is used on continues dataset ---2 types of Regression 1) simple Regression ----------Linear simple Regresssion --------Non Linear simple Regression

--If one feature is mapped for target then it is simple Regression if this selected feature and target have linear relationshipthen it is called as simple linear Regression ---if it is non liearity between feature and targetthen it is called as simple non linear regression

2)Multiple Regression --------Linear multiple Regression ------Non-linear multiple Regression

if more than one feature are mapped with targetthen it will be multiple Regression

if this selected feature and target have linear relationshipthen it is called as multiple linear Regression ---if it is non liearity between feature and targetthen it is called as multiple non linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
x=np.array([1,2,3,4])
slope=0.5
intercept=4
y=slope *x+ intercept
y
```

```python
plt.plot(x,y)
plt.scatter(x,y)
plt.show()
```

```python
area=np.array([600,800,1000,1200,1500,2000])  #---actual x
price=np.array([4,6,9,12,16,21])  #---actual y
plt.scatter(area,price)
```

```python
ypred=area*slope+intercept
ypred
```

```python
area_m=np.mean(area)
price_m=np.mean(price)
slope=np.sum((area-area_m)*(price-price_m))/np.sum((area-area_m)**2)
intercept=price_m-(slope*area_m)
```
