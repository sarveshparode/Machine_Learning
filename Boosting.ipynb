```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data=pd.read_csv("bank.csv")
data.head()
```

```python
#define training and testing data
X=data.drop('deposit',axis=1)
y=data['deposit']
```

```python
#train test split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
#fitting of weak classifier
#weal learnar of weak classifier mean the alogorithm that consist of error or mistake
dt_clf=DecisionTreeClassifier(max_depth=1,random_state=0)
dt_clf.fit(xtrain,ytrain)
dt_score=dt_clf.score(xtest,ytest)
print("decision tree score:",dt_score)
```

```python
#fitting of weak classifire with adaboost
from sklearn.ensemble import AdaBoostClassifier
ada_clf=AdaBoostClassifier(base_estimator=dt_clf,random_state=0)
#n_estimaters= it controls the number of weak learners
#based estimater--- it helps to specify different ML algorithm
ada_clf.fit(xtrain,ytrain)
ada_score=ada_clf.score(xtest,ytest)
print("ada boost score is",ada_score)
```

```python
#fitting of weak classifire with Gradient Boosting
from sklearn.ensemble import GradientBoostingClassifier
gb_clf= GradientBoostingClassifier(random_state=0)

gb_clf.fit(xtrain,ytrain)
gb_score=gb_clf.score(xtest,ytest)
print("Gradient  boost score is",gb_score)
```

```python
#fitting of weak classifire with XtreamGradient Boosting
from xgboost import XGBClassifier
xgb_clf= XGBClassifier(base_estimators=dt_clf,random_state=0)

xgb_clf.fit(xtrain,ytrain)
xgb_score=xgb_clf.score(xtest,ytest)
print("XGBradient  boost score is",xgb_score)
```
